```python
# Get geo data from .csv files
css_df = pd.read_csv(data_path+'geo_info_css.csv')
traveller_df = pd.read_csv(data_path+'geo_info_travellers.csv')
poi_df = pd.read_csv(data_path+'geo_info_POIs.csv')
```

### Carsharing system network built in Copenhagen Denmark
* Green nodes: carsharing stations
* Blue nodes: points of interest
* Curved lines: generated trips

```python
import folium
# Make an empty map
map = folium.Map(location=[55.68,12.57], tiles='cartodbpositron', zoom_start=12)
 
# add marker for css
for i in range(len(css_df)):
    icon_settings = {'icon': 'car', 'icon_shape': "marker", 'border_color': "green",
                     'border_width': 2, 'icon_size': [23,23], 'background_color': 'rgba(255,255,255,0.7)',
                     'text_color': 'black', 'spin': False}
    beautify_icon = BeautifyIcon(**icon_settings)
    marker = folium.Marker(location=[css_df.iloc[i]['lat'], css_df.iloc[i]['lng']], 
                           tooltip=css_df.iloc[i]['css_title'], icon=beautify_icon)
    marker.add_to(map)

# add marker for POI
for i in range(len(poi_df)): 
    folium.Circle(
      location=[poi_df.iloc[i]['POI_lat'], poi_df.iloc[i]['POI_lng']],
      popup=poi_df.iloc[i]['POI_name'],
       radius = 20,
      color="#3186cc",
      fill=True,
      fill_color="#3186cc"
).add_to(map)


# Add latitude and longitude grid lines
folium.LatLngPopup().add_to(map)
# Add a MiniMap to display an overview with a scale bar
minimap = MiniMap(toggle_display=True)
map.add_child(minimap)

display(map)
```

```python
# add curved lines for travellers' trips
# randomly pick 50 travellers
import random
traveller_num = 200
my_list = list(range(0, len(traveller_df)))
random_subset = random.sample(my_list, traveller_num)

for i in random_subset:
    lat_o = traveller_df.iloc[i]['lat_o']
    lng_o = traveller_df.iloc[i]['lng_o']
    lat_d = traveller_df.iloc[i]['lat_d']
    lng_d = traveller_df.iloc[i]['lng_d']
    point1_coords = [lat_o, lng_o]
    point2_coords = [lat_d, lng_d]
    folium.PolyLine(
    locations=[point1_coords, point2_coords],
    color="darkblue",
    opacity=0.8,
    weight=0.5,
    delay=800
).add_to(map)

display(map)
```
